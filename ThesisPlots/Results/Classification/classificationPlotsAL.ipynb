```python
# simulate and plot a spike-and-slab distribution
import scipy.stats as stats
import matplotlib.pyplot as plt

# set the color
colors = sns.color_palette("viridis", 8)

# loop through the colors and assign colors_[i] to the color
for i in range(8):
    exec(f'color_{i} = colors[i]')

# set the style
sns.set_style("whitegrid")

# set the figure size
plt.figure(figsize=(10, 6))

# set the x values
x = np.linspace(-10, 10, 1000)

# set the spike and slab distribution
y = 0.5 * stats.norm.pdf(x, 0, 1) + 0.5 * stats.norm.pdf(x, 0, 3)

# set the slab distribution
y_slab = 0.5 * stats.norm.pdf(x, 0)


# plot the spike and slab distribution
plt.plot(x, y, color=color_0)
plt.fill_between(x, y, color=color_0, alpha=0.5)
plt.plot(x, y_slab, color=color_3)

plt.xlabel('x')
plt.ylabel('Density')
plt.title('Spike-and-Slab Distribution')

# horizontal arrow pointing to the spike
plt.annotate('Spike', xy=(0, 0.25), xytext=(5, 0.3),
                arrowprops=dict(facecolor='black', shrink=0.05))
```

```python
"""Plot a spike-and-slab distribution. The spike-and-slab distribution is a mixture of a spike distribution and a slab distribution. The spike distribution is a point mass at zero, and the slab distribution is a normal distribution centered at zero. The spike-and-slab distribution is a useful prior for Bayesian variable selection.
set the variance of the slab to zero to emulate mc dropout
"""

# set the style
sns.set_style("whitegrid")

# set the figure size
plt.figure(figsize=(10, 6))

# set the x values
x = np.linspace(-10, 10, 1000)

# set the spike and slab distribution
y = 0.5 * stats.norm.pdf(x, 0, 0) + 0.5 * stats.norm.pdf(x, 0, 3)

# set the slab distribution
y_slab = 0.5 * stats.norm.pdf(x, 0, 3)

# plot the spike and slab distribution
plt.plot(x, y, color=color_0)
plt.fill_between(x, y, color=color_0, alpha=0.5)
plt.plot(x, y_slab, color=color_3)

plt.xlabel('x')
plt.ylabel('Density')
plt.title('Spike-and-Slab Distribution with Zero Variance')
```
